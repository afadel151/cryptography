# Chiffrement de César
Le chiffre de César ou le code de César est un chiffrement par décalage. C'est une méthode de chiffrement très simple utilisée par Jules César dans ses correspondances secrètes. Dans cette méthode, le texte chiffré s'obtient en remplaçant chaque lettre du texte clair original par une lettre à distance fixe, toujours du même côté, dans l'ordre de l'alphabet. Pour les dernières lettres, on reprend au début. 

Par exemple avec un décalage de 3 vers la droite, A est remplacé par D, B devient E, et ainsi jusqu'à W qui devient Z, puis X devient A etc. 
La longueur du décalage, 3 dans l'exemple évoqué, constitue la clé du chiffrement qu'il suffit de transmettre au destinataire pour que celui-ci puisse déchiffrer le message.
 


## Chiffrement
Écrivez une fonction qui chiffre un texte en anglais avec l'algorithme de César. La clé et le texte à chiffrer seront passés en paramètres.

```python
def Encrypte(Text,key):
    EncryptedText=''
    for i in range(len(Text)):
        char = Text[i]
        if (char.isupper()):
            EncryptedText += chr((ord(char) + key-65) % 26 + 65)
        elif char.islower():
            EncryptedText += chr((ord(char) + key - 97) % 26 + 97)
    return EncryptedText

print(Encrypte('Fadel',3))
```

## Déchiffrement
Écrivez une fonction permettant le déchiffrement d'un texte en anglais qui est déja chiffrer avec l'algorithme de César. La clé et le texte à déchiffrer seront passés en paramètres.


```python
def Decrypte(EncryptedText,key):
    Text=''
    for i in range(len(EncryptedText)):
        char = EncryptedText[i]
        if (char.isupper()):
            Text += chr((ord(char)-key-65)%26 + 65)
        elif (char.islower()):
            Text += chr((ord(char)-key-97)%26 + 97)
    return Text

print(Decrypte('Idgho',3))
```

## Cryptanalyse de cet algorithme

### Attaque par force brute
Une première méthode est appelée une attaque par force brute.
Comme il n'y a qu'un nombre limité de décalages (vingt-six dont un inutile), il suffit de tester tous les chiffrements possibles jusqu'à trouver le bon. 
Une méthode simple pour mener l'attaque est de prendre un fragment du texte chiffré et d'écrire dans un tableau tous les décalages possibles.
Par conséquent, il est demandé d'écrire une fonction pour décrypter le message suivant:

`Rdcvgpijapixdc ndj hjrrthhujaan qgtpz Rtphtg rxewtg`

```python
def BruteForce(EncryptedText):
   for i in range(26):
      print(str(i) + ":" +Decrypte(EncryptedText,i))

BruteForce('Rdcvgpijapixdc ndj hjrrthhujaan qgtpz Rtphtg rxewtg') 
# key = 15
```

### Attaque par analyse de fréquences
Une deuxième méthode pour déterminer la clé k du chiffre de César est la suivante. 
On calcule, pour chaque valeur de k, la fréquence d'apparition de chacune des lettres dans le message chiffré.
À l'aide d'un texte de référence, on détermine la fréquence d'apparition de chacune des lettres en anglais. 
On choisit la clé qui permet d'obtenir une liste de fréquences aussi proche que possible de la liste de référence. 
Pour cela, si $f = (f_0, . . . , f_{25}) $ et $g = (g_{0}, . . . , g_{25})$ sont deux listes de fréquences, on détermine leur distance via la fonction
\begin{equation}
d(f,g)= \sum_{i=0}^{25} | f_i - g_i|
\end{equation}

Par conséquent, il est demandé d'écrire un programme pour décrypter le message ce trouvantt dans le fichier ``TextChiffe.txt''.

```python
# la fréquence d'apparition de chacune des lettres en anglais
frequencies= [0.082,0.015,0.028,0.043,0.13,0.022,0.02,0.061,0.07,0.0015,0.0077,0.04,0.024,0.067,0.075,0.019,0.00095,0.06,0.093,0.091,0.028,0.0098,0.024,0.0015,0.02,0.00074]
def FreqAnalysis(EncryptedText):
   dict = {}
   text = str(EncryptedText)
   for i in range(len(text)):
      ch = text[i].lower()
      if ch.isalpha():
         if(dict.get(ch) != None):
            dict[ch] += 1
         else:
            dict[ch] =1
   counts = [0] * 26
   total = 0
   for letter, cnt in dict.items():
        idx = ord(letter) - 97
        if 0 <= idx < 26:
            counts[idx] = cnt
            total += cnt
   if total == 0:
        return 0
      
   key=0
   best_key = 0
   best_score = float('inf')
   for key in range(26):
         rotated = [counts[(i + key) % 26] for i in range(26)]
         score = 0.0
         for i in range(26):
            expected = frequencies[i]*total
            if expected > 0:
               score += ((rotated[i] - expected) ** 2) / expected
         if score < best_score:
            best_score = score
            best_key = key
   return best_key

```

```python
with open('TextChiffer1.txt') as f:
    EncryptedText = f.readlines()
    key = FreqAnalysis(EncryptedText)
    print(key)
    message = Decrypte(str(EncryptedText),key)
    print(message)
```
